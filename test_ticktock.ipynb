```python
from Imagehandler import Imagehandler
import yaml
import glob
import os
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
import sys
%matplotlib inline
```

```python
os.chdir('/content/off_datatest/qr_data_noOFF/')
```

```python
files = [f for f in os.listdir('./') if os.path.isfile(f)]
```

```python
for file in files:
  print('---------File: '+file+'---------')
  image = cv.imread(file)
#cv.imshow('Qr code is ',image)
  h,w = image.shape[:2]
  print(image.shape)
  plt.imshow(image,cmap='gray')
  plt.axis('off')
  plt.show()
  print('Locking for a QR code...')
  try:
    obj = Imagehandler(image)
    TransformImage = obj.QRCodeInImage()
    im = TransformImage
    plt.imshow(im)
    plt.axis('off')
    plt.show()
  except:
    print('This file has an issue during QR detection: '+file)
  
  
```
